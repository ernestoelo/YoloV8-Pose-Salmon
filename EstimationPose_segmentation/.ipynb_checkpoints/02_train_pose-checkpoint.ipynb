```python
#librer√≠as ---
from ultralytics import YOLO
import torch
from IPython.display import Image, display
import os
```

```python
#Verificar entorno y dispositivo ---
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"‚úÖ Entrenamiento en: {device}")
```

```python
# ---Configuraci√≥n de rutas ---
DATA_YAML = "./data.yaml"     # archivo de configuraci√≥n del dataset
MODEL_BASE = "yolov8n-pose.pt" # modelo base liviano
PROJECT_DIR = "runs/pose"      # carpeta donde se guardar√°n los resultados
MODEL_NAME = "salmon_pose"     # nombre del experimento
```

```python
# --- üß† 5. Cargar modelo base YOLOv8 Pose ---
model = YOLO(MODEL_BASE)
```

```python
# --- üöÄ 6. Entrenamiento ---
results = model.train(
    data=DATA_YAML,        # dataset COCO keypoints
    epochs=100,            # n√∫mero de √©pocas
    imgsz=640,             # tama√±o de entrada
    batch=8,               # tama√±o del batch
    device=device,         # usar GPU si est√° disponible
    workers=4,
    optimizer="AdamW",
    lr0=0.001,
    project=PROJECT_DIR,
    name=MODEL_NAME,
    exist_ok=True,
    cache=True,
    patience=10
)
```
